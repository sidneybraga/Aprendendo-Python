Crie uma função que receba uma frase e retorne quantas palavras tem a frase. Dica: Utilize o método split().

```python
def numeropalavras(frase):
  lista=(frase.split(' '))
  npalavras=len(lista)
  return npalavras
```

***Exercicio 2.*** \\
Crie uma função que lê um arquivo de extensão .txt e retorne quantas palavras tem esse arquivo. Dica: utilize o método readlines().
***Minha solução:***

```python
def numeropalavrastxt(file_path):
  with open(file_path, encoding= 'latin-1') as f:
    f_data = f.readlines()
    f_data_lista = []
    numerodepalavras = 0
  for i in range(len(f_data)):
    f_data_lista.append(f_data[i].split(' '))
    numerodepalavras += len(f_data_lista)
  return numerodepalavras
```

```python
numeropalavrastxt('/content/PIB_100_maiores_cidades_2017.txt')
```

***Solução professor:***

```python
def numero_de_palavras(arquivo_path):
  with open(arquivo_path, encoding= 'latin1') as f:
    string = f.read()
    f.close
  lista_palavras= string.replace('\n','')
  lista_palavras = lista_palavras.split(' ')
  numero_palavras = len(lista_palavras)
  return numero_palavras
```

```python
numero_de_palavras('/content/PIB_100_maiores_cidades_2017.txt')
```

***Resumo:***
A função ***.read()*** retorna todo o arquivo txt como uma unica *string*. A função
***.replace('\n','')*** limpa a string substituindo todos os valores \n por um espaço nulo. A função ***.split(' ')*** cria uma lista de palavras separadas por espaços no arquivo txt. E por fim a função ***len()*** retorna o tamanho dessa lista, e portanto, o numero de palavras.

***Desafio 1 - Plotlando Coordenadas*** \\
Crie um módulo que leia um arquivo de pontos 3D e faça um gráfico desses pontos.

Arquivo de pontos: PONTOS_ESFERA.txt

Fonte para fazer o gráfico: https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html \\
**Minha Resolução:**

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib inline
x=[1,2,3,4,5]
y=[1,2,3,4,5]
z=[1,2,3,4,5]
fig = plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x,y,z)
```

```python
with open('/content/PONTOS_ESFERA.txt') as f:
    f_data = f.readlines()
    f_data_linhas = []
    f_coordenadas_string = []
    x_f= []
    y_f= []
    z_f= []
    for i in range(len(f_data)):
      s = f_data[i].replace('\n','')
      f_data_linhas.append(s.split(','))
for k in range(len(f_data)):
  f_coordenadas_string.append(f_data_linhas[k][3:])
  x_f.append(float(f_coordenadas_string[k][0]))
  y_f.append(float(f_coordenadas_string[k][1]))
  z_f.append(float(f_coordenadas_string[k][2]))
fig = plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x_f,y_f,z_f)
```

***Resumo:*** \\
A função ***.readlines()*** transforma cada linha do arquivo **.txt** em elementos de uma lista **f_data**. O ***for*** percorre essa lista de elementos. O método ***.replace('\n','')*** retorna uma *string* de cada elemento da lista, e armazena essa string em **s**. O método ***split(',')*** separa a string **s** em virgulas e armazena em uma lista **f_data_linhas**. O novo **for** em **k** percorre a lista **f_data_linhas** a partir do elemento 3 e armazena em uma lista de coordenadas ***f_coordenadas_string***. No entanto, essa lista de coordenadas esta no formato de *string*, então as listas **x_f, y_f e z_f** transformam os elementos string **x y e z** da lista de coordenadas em ***floats***. Assim, cria-se a lista de coordenadas desejada e plota-se o grafico em ax.scatter(x_f,y_f,z_f)

**Solução Professor:**

```python
def pega_coordenadas(file_name):
  with open(file_name) as f:
      f_data = f.readlines()
      f.close

  linhas = []

  for i in range(len(f_data)):
    linhas.append(f_data[i].split(','))

  coordenadas = {}
  coordenadas['x'] = []
  coordenadas['y'] = []
  coordenadas['z'] = []
  for i in range(len(linhas)):
    x=float(linhas[i][3])
    y=float(linhas[i][4])
    z=float(linhas[i][5])

    coordenadas['x'].append(x)
    coordenadas['y'].append(y)
    coordenadas['z'].append(z)

  return coordenadas

def plot_3dcoord(coord_dict):
  import matplotlib.pyplot as plt
  from mpl_toolkits import mplot3d
  %matplotlib inline
  fig = plt.figure()
  ax= fig.add_subplot(111, projection = '3d')

  x = coord_dict['x']
  y = coord_dict['y']
  z = coord_dict['z']

  ax.scatter(x,y,z)
```

```python
coordenadasesfera = pega_coordenadas('/content/PONTOS_ESFERA.txt')
```

```python
plot_3dcoord(coordenadasesfera)
```
